## **Results**

```python
from helpers.metrics import get_results, print_results
```

```python
datasets_results= ["nba"]
split_path = "../data/splits"
output_folder = "../data/output"
csv_out = f"{output_folder}/MORAL_edge_distributions.csv"
k = 1000

results_combined = {}
res_base = get_results(
    datasets=datasets_results,
    folder=output_folder,
    graph_stats_csv=csv_out,
    splits_dir=split_path,
    k=k
)

if res_base:
    for ds, metrics in res_base.items():
        results_combined["nba 2Layer"] = metrics
        
for i in range(3, 9):
    res_i = get_results(
        datasets=datasets_results,
        folder=output_folder,
        graph_stats_csv=csv_out,
        splits_dir=split_path,
        k=k,
        model_string=f"_MORAL_{i}HOP_GAE",
    )

    for ds, metrics in res_i.items():
        results_combined[f"{ds} {i}Layer"] = metrics

results = results_combined
print_results(results, k)
```
